```python
relevant_cols = df[["date", 'hour', 'day', 'quantity'
       ,'month', 'week', 'Quarter', 'location', 'Volume', 'clean_product_name']]
relevant_cols = relevant_cols[relevant_cols["location"] == "Jaksel"]
```

```python
order_quantity = relevant_cols.groupby(["date", "clean_product_name", 'day','month', 'week', 'Quarter'])["quantity"].sum().reset_index()
```

```python
kawa_kawa_orders = order_quantity[order_quantity["clean_product_name"] == "kawa kawa anggur hijau"]
```

```python
periods = 365 * 1
dates = pd.date_range(start='2024-01-01', periods=periods)
holidays_df = id_holidays(dates)
holidays_df.tail()
```

```python
holidays_df = holidays_df.reset_index()
```

```python
holidays_df
```

```python
holidays_df['index'] = pd.to_datetime(holidays_df['index'])
kawa_kawa_orders['date'] = pd.to_datetime(kawa_kawa_orders['date'])

# Merge the DataFrames
merged_df = holidays_df.merge(kawa_kawa_orders, left_on='index', right_on='date', how='inner')
```

```python
train_df_complete = fill_gaps(merged_df, freq='D', time_col="date", id_col="date")
```

```python
forecast_df = nixtla_client.forecast(
    df=merged_df,
    h=14,
    time_col='index',
    target_col="quantity"
)

# Plot predictions
# nixtla_client.plot(
#     df=merged_df, 
#     forecasts_df=forecast_df, 
#     time_col='timestamp', 
#     target_col='value'
# )
```
